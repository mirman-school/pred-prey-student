This is useful because we can track population over time in a list. Instead of a simple number like `bacteria`, let's try it with a list. We start with a single item in the list, our starting population. Then we `append()` to the list the doubled amount.

**Fill in the missing code below and run the cell**

````python
bacteria_population = [10]

days = 10

for i in range(days):
    last_pop = bacteria_population[-1] # Get the last bacteria population
    next_pop = last_pop * 2 # double the bacteria is the next step (FIX THIS LINE)
    # Below this line, use append() to add next_pop to the bacteria_population list
    bacteria_population.append(next_pop)
    
print(bacteria_population)
````

### Natural Death Rate
Let's say 10% of the population dies every day from natural causes. How can we include that in our calculation?

Think back to our population growth loop:

```python
for i in range(days):
    last_pop = bacteria_population[-1] # Get the last bacteria population
    next_pop = last_pop * 2 # double the bacteria for the next step (FIX THIS LINE)
    bacteria_population.append(next_pop)
```
Rather than just multiplying `last_pop` by 2, we need to account for the death rate. Let's try to state the growth in English first:

> The bacteria population grows every day by doubling, but decreases by the percentage of the population that dies every day.

"Grows every day" indicates addition or multiplication. "Decreases" indicates a subtraction.

So if we have the death rate as a decimal value:

```python
death_rate = .1 # 10% daily death rate
new_pop = (last_pop * 2) - (last_pop * death_rate)
```

**Let's try it now**

````python
bacteria_population = [10]
death_rate = .5
days = 10

for i in range(days):
    last_pop = bacteria_population[-1] # Get the last bacteria population
    next_pop = (last_pop * 2) - (last_pop * 1) # Use the death rate in the second term
    bacteria_population.append(next_pop)
    
x_range = arange(0,11,1)
f = pl.figure(1) # start the figure
pl.plot(x_range,bacteria_population)
pl.xlabel("") # Set a proper x label
pl.ylabel("") # Set a proper y label
pl.title("") # Set a proper graph title
pl.show(f) # Actually show the graph
````

## Lotka-Volterra

Two scientists, decades apart, thought about these same issues. Volterra, building on Lotka's earlier work, finalized a model that looks like this: 

![l-v](https://wikimedia.org/api/rest_v1/media/math/render/svg/79752d662d4760abcc84c6f0bb94d708f17ff442)

* α is the natural growth rate of prey in the absence of predation
* β is the death rate per encounter of prey due to predation
* σ is related to the growth rate of predators
* γ is the natural death rate of predators in the absence of food (prey)

Now explain to your neighbor what these things mean, and what you think the equations suggest.

We're going to make a function that does L-V for us. But we'll name the variables in useful ways.

**This is not code you have to complete. But you should change the numbers to see how the populations respond.**

**As you alter the parameters of the code, discuss with your neighbor what you're doing and _why_ you're doing it. Also, make a prediction about how your change will alter the graph.**

````python
def lotka_volterra(state,t):
  prey_pop = state[0] # DON'T TOUCH
  pred_pop = state[1] # DON'T TOUCH
  prey_growth_rate = 0.1 # Change
  predation_rate =  0.1 # Change
  pred_growth_rate = 0.1 # Change
  pred_death_rate = 0.1 # Change
  prey_change = prey_pop*(prey_growth_rate - predation_rate*pred_pop) # DON'T TOUCH
  pred_change = -pred_pop*(pred_death_rate - pred_growth_rate*prey_pop) # DON'T TOUCH
  return [prey_change,pred_change] # DON'T TOUCH

t = arange(0,500,1) # Change the middle number for longer/shorter simulations
state0 = [0.5,0.5] # Change if you want to alter the starting situation
state = odeint(lotka_volterra,state0,t) # DON'T TOUCH
pl.figure(1, figsize=[20,10])
pl.plot(t,state)
pl.ylim([0,3]) # Change if you want the y axis to grow/shrink
pl.xlabel('Time')
pl.ylabel('Population Size')
pl.legend(('Prey','Predator'))
pl.title('Lotka-Volterra equations')
pl.show(1)
````
